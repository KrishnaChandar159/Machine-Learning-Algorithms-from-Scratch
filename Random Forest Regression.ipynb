```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
data = pd.read_csv('air_quality.csv') 
data.head()
```

```python
#Checking out the summary
#We couls see some NaN values as the count does'nt add up
data.describe()
```

```python
data['wind_direction'].notnull().sum()
```

```python
#Creating a sub DataFrame for the where temperature, pressure and rain is NaN
data_20 = data[data['temperature'].isnull()]
data_20
```

```python
#filling nan values with mean of temp,pres,rain on that particular day
for i in range(len(data_20.index)):
    data_sub = data[(data['year'] == data_20.iloc[i,0]) & (data['month'] == data_20.iloc[i,1]) & (data['day'] == data_20.iloc[i,2])]
    data_sub_1 = data_sub.iloc[:,[5,6,7,9]]
    data_20.iloc[i,[5,6,7,9]] =(data_sub_1.mean(axis=0,skipna = True,numeric_only = True)).to_numpy()
```

```python
data[data['temperature'].isnull()] = data_20
```

```python
#removing the -1 and nan PM2.5 values
no_winddir = data
no_winddir = no_winddir[(no_winddir['PM2.5'] != -1) & (no_winddir['PM2.5'].notnull())]
no_winddir.head()
```

```python
#creating a new variables which combines 'year','month','day' to a single feature by keeping the count of days
no_winddir['day_sno'] = (no_winddir['year'] - 2013)*365 + (no_winddir['month']-3)*30 + (no_winddir['day'])
no_winddir.head()
```

```python
#using the wind_direction in terms of four features representing the N,W,S,E count in each row 
N = []
S = []
E = []
W = []
for i in no_winddir['wind_direction']:
    N.append(str(i).count('N'))
    S.append(str(i).count('S'))
    W.append(str(i).count('W'))
    E.append(str(i).count('E'))
no_winddir['N'] = N
no_winddir['S'] = S
no_winddir['W'] = W
no_winddir['E'] = E
no_winddir.head()
```

```python
#removing year,month,day and wind_direction
no_winddir.drop(["year", "month", "day","wind_direction"], axis = 1, inplace = True)
no_winddir.head()
```

```python
no_winddir.describe()
```

```python
#Normalization doesnt provide the best results. so the data is left unnormalized
```

```python
#train test split
X = no_winddir.iloc[:,[0,2,3,4,5,6,7,8,9,10]].values.astype(float)
y = no_winddir.iloc[:,1].values.astype(float)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Fitting Random Forest Regression to the dataset 
# import the regressor 
from sklearn.ensemble import RandomForestRegressor 
# create regressor object 
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0) 
# fit the regressor with x and y data 
regressor.fit(X_train, y_train)
```

```python
#predicting test_data rmse
y_pred = regressor.predict(X_test)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
rmse
```

```python
#predicting train_data rmse
y_pred_1 = regressor.predict(X_train)
from sklearn.metrics import mean_squared_error
mse_1 = mean_squared_error(y_train, y_pred_1)
rmse_1 = np.sqrt(mse_1)
rmse_1
```

```python
#preprocessing the input data for predictions
X_tbp = data[data['PM2.5']==-1]
X_tbp['day_sno'] = (X_tbp['year'] - 2013)*365 + (X_tbp['month']-3)*30 + (X_tbp['day'])
N = []
S = []
E = []
W = []
for i in X_tbp['wind_direction']:
    N.append(str(i).count('N'))
    S.append(str(i).count('S'))
    W.append(str(i).count('W'))
    E.append(str(i).count('E'))
X_tbp['N'] = N
X_tbp['S'] = S
X_tbp['W'] = W
X_tbp['E'] = E
X_tbp.head()
X_tbp.drop(["year", "month", "day","wind_direction"], axis = 1, inplace = True)
X_tbp.head()
```

```python
#predictions
y_tbp = regressor.predict(X_tbp.iloc[:,[0,2,3,4,5,6,7,8,9,10]])
y_tbp
```

```python
#converting the output predictions to a csv file
y_tbp_df = pd.DataFrame(y_tbp)
y_tbp_df.to_csv("output.csv",index = False)
```
