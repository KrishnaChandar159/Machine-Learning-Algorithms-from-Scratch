# Application of Convolution
Executing a python program to perform convolution on the 2 images with given 3x3 filter and display the results i.e. filtered images of Cat and Dog.

Filter given as
    
    1 0 0
    0 1 0
    0 0 1

Pseudo Code

    Create 2 classes Cat and Dog

    Cat and Dog classes have readCatImg() and readDogImg() functions respectively

    readCatImg() : reads an image from system in vector
    readDogImg() : reads an image from system in vector

    Create 3 class Convolution inherits Cat, Dog

    Call readCatImg() & readDogImg() functions
    
    Now lets do element-wise multiplication (convolution) with filter on read images.Store results in vectors.Convert vectors to image

    Display image.

```python
import matplotlib.image as mimg
import matplotlib.pyplot as plt
import numpy as np
class Cat:
    def readCatImg():
        return mimg.imread('cat.jpeg')
class Dog:
    def readDogImg():
        return mimg.imread('dog.jpg')
class Convolution(Cat,Dog):
    def readImg(self):
        array = []
        array.append(Dog.readDogImg())
        array.append(Cat.readCatImg())
        for i in array:
            self.convolve(i)
    def convolve(self,arr):
        #arr = Cat.readCatImg()
        (m,n,size) = arr.shape
        s=1
        output_len = len(range(0,m-2,s))
        input_len = len(range(0,n-2,s))
        filter_mat = np.array([[1,0,0],[0,1,0],[0,0,1]])
        final = []
        input_mat = arr
        for i in range(0,m-2,s):
            for j in range(0,n-2,s):
                new_arr = []
                new_arr = input_mat[i:i+3,j:j+3]
                count = np.array([0,0,0])
                for k in range(3):
                    for l in range(3):
                        count = count + new_arr[k,l]*filter_mat[k,l]
                final.append(list(count))
        final = [final[i:i+input_len] for i in range(0,output_len*input_len,input_len)]
        plt.imshow(np.array(final))
        plt.show()
obj = Convolution()
obj.readImg()
```
